```python
age=[21,3,543,34,54,67,34,23,12,45,765]
plt.boxplot(age)
plt.show()
```

```python
#in this data there is 2 outliers
```

```python
#if we want to remove them its on the data- if the data is like football data where anomly is ronaldo messi etc.
#they are better than other players exceptionally - hence we can't remove them
#we can remove outliers when the data is about credit card sm, mobile etc. 
#hence it depends on data either we have to drop or not (work with outliers)
```

```python
#Outliers Removal ->
```

```python
#Inter Quartile Range (IQR) ->
#IQR = Q3 - Q1
```

```python
#Step to remove outlier
#1. Calculate Q1(25 percentile) and Q3(75 percenntile)
#2. Calculate the IQR = Q3-Q1
#3. Calculate the lower Bound -> Q1 - 1.5IQR
#4. Upper Bound -> Q3 + 1.5IQR
#5. Any value that's below lower bound and above upper bound - we have to remove it!
```

```python
#the box remaining is IQR
#upper bound as UB shown in fig, same as lower bound LB 
#we removing lower to lower bound and upper to upper bound so we get normal data!
```

```python
#Step 1 ->
Q1 = np.percentile(age,25)
Q3 = np.percentile(age,75)
```

```python
#Step 2 ->
IQR = Q3 - Q1
print(IQR)
```

```python
#Step 3 ->
lower_bound= Q1 - 1.5*IQR
#Step 4 ->
upper_bound= Q3 + 1.5*IQR
```

```python
#Step 5 ->
normal_age=[]
for i in age:
    if i<lower_bound or i>upper_bound:
        print("The outlier(s) is :", i)
    else:
        normal_age.append(i)
```

```python
normal_age
```

```python
#outliers are removed! yay! - there is no excess points in data
```
