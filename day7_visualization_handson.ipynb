```python
#create a headmap for entire data to study correlations between each of the columns
df= pd.read_csv('cars.csv',skipinitialspace=True) #skipinitialspace means int wont change to obj due to extra spacing
numData = df.select_dtypes([int,float]) #because we needed int data and float data for that we have requried skipinitialspace
corrMat = numData.corr() #correlate matrix it means- corr is used for correlated -!
corrMat
```

```python
sns.heatmap(corrMat)
plt.show()
```

```python
sns.heatmap(corrMat,annot=True) #it is a parameter used to show the value in heatmap
plt.show()
```

```python
sns.heatmap(corrMat,annot=True,cmap='Set2') #it is a parameter used to show the value in heatmap
plt.show()
```

```python
sns.heatmap(corrMat,annot=True,cmap='Blues') #it is a parameter used to show the value in heatmap
plt.show()
```
