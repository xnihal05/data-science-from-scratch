```python
#Create a disct with values in list and create it to dataframe
dic = {
    'Name' : ['Yashmeen', 'Nihal', 'N/A'],
    'Age' : [19, 20, 0],
    'Adress' : ['abc' , 'def' , 'ijk']
}
dic_to_dataframe = pd.DataFrame(dic)
dic_to_dataframe
```

```python
#Rename Name to Student Name
dic_to_dataframe.rename(columns={"Name":"Student Name"}, inplace =True)
dic_to_dataframe
```

```python
#All column where age is greater than and equal to 19
age_greater = dic_to_dataframe[dic_to_dataframe['Age']>=19]
age_greater
```
