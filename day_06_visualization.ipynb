```python
#Line PLot --> just like stocks plot , sales etc...( For continuous data )
```

```python
days = [ 'sun', 'mon', 'tue','wed','thu','fri','sat']
sales = [100,50,150,50,400,432,130]
```

```python
import matplotlib.pyplot as plt
days = [ 'sun', 'mon', 'tue','wed','thu','fri','sat']
sales = [100,50,150,50,400,432,130]
plt.xlabel('Days in the week')
plt.ylabel('Sales in the week')
plt.title('Sales  v/s  Week')
plt.plot(days,sales, color='red', marker='o') # its like (x,y)
#marker o means it'll give o points where out edge is - we can also give '*', it'll look * as a marker, v is also there
plt.show()
```

```python
plt.title('Sales  v/s  Week')
plt.plot(days,sales, color='red', marker='o',linestyle='dashed') # its like (x,y)
#marker o means it'll give o points where out edge is - we can also give '*', it'll look * as a marker, v is also there
plt.show()
```

```python
plt.xlabel('Days in the week')
plt.ylabel('Total Sales')
plt.title('Sales  v/s  Week')
plt.plot(days,sales, color='red', marker='v',linestyle='dashed', markersize= 15, markerfacecolor='yellow', markeredgecolor='blue') 
# its like (x,y) 
plt.grid()
#marker o means it'll give o points where out edge is - we can also give '*', it'll look * as a marker, v is also there
plt.show()
```

```python
#grid just make accuracy and descreases the possibility of errors
```

```python
days = [ 'sun', 'mon', 'tue','wed','thu','fri','sat']
apple_sales = [1030,150,1501,530,400,432,130]
pineapple_sales = [1300,500,150,50,200,632,1230]
desi_apple_sales = [60,40,456,678,142,234,1330]
plt.plot(days,apple_sales, color='red',label ='apple', marker='o', markerfacecolor='yellow', markeredgecolor='blue')
plt.plot(days,pineapple_sales, color='brown',label ='pineapple', markerfacecolor='orange', markeredgecolor='red')
plt.plot(days,desi_apple_sales, color='orange',label ='desi apple',markerfacecolor='red', markeredgecolor='black')
plt.legend()
plt.grid()
plt.show()
```

```python
#Bar Plot - when you want to see the data as well (line plot is generally used for big data)
# when we want to compare between catogeries then we will use bar plot simply !
```

```python
days = [ 'sun', 'mon', 'tue','wed','thu','fri','sat']
sales = [1030,150,1501,530,400,432,130]
plt.bar(days,sales, color=(['blue','red','indigo','yellow','black','cyan','purple']), width=0.6)
plt.title('Days VS Sale')
plt.grid(color='black')
plt.show()
```

```python
#there is another graph we can define from bar which is 'barh' = bar horizontally (y axis,x axis)
```

```python
days = [ 'sun', 'mon', 'tue','wed','thu','fri','sat']
sales = [1030,150,1501,530,400,432,130]
plt.barh(days,sales, color=(['blue','red','indigo','yellow','black','cyan','purple']))
plt.title('Days VS Sale')
plt.show()
```

```python
#pi chart - it will show the proportion(percentage) of the catogeries 
value=[10,30,50,80]
cato = ['mango','apple','banana','strawberry']
plt.pie(value,autopct='%1.1f%%',labels=cato) 
plt.title('pi-chart')
plt.show()
```

```python
# first % and last % is just like a syntax and inside we have to define how  value we  want to define as float
```

```python
# giving f only makes it more value after decimal= if we want only one value after point then we use 1.1 and f means float
value=[10,30,50,80]
cato = ['mango','apple','banana','strawberry']
plt.pie(value,autopct='%1f%%',labels=cato) 
plt.title('pi-chart')
plt.show()
```

```python
value=[10,30,50,80]
cato = ['mango','apple','banana','strawberry']
plt.pie(value,autopct='%1.2f%%',labels=cato) 
plt.title('pi-chart')
plt.show()
```

```python
value=[10,30,50,80]
cato = ['mango','apple','banana','strawberry']
plt.pie(value,autopct='%i%%',labels=cato) 
plt.title('pi-chart')
plt.show()
```

```python
#i% just means percentage in integer
```

```python
#explode = makes the one part to the outside
value=[10,30,50,80]
cato = ['mango','apple','banana','strawberry']
plt.pie(value,autopct='%1.1f%%',labels=cato, explode=[0,0,0,0.2], shadow=True)  #shadow just makes shadow for looks
plt.title('pi-chart')
plt.show()
```

```python
#Scatter Plot -> if am removing lines from line plot we got dots only thats similar to scatter plot (rough idea)
roll_no=[1,2,3,4,5,6,7,8]
marks=[60,70,80,76,55,45,34,67]
plt.scatter(roll_no,marks,color='red')
plt.xlabel('Roll Number of Students')
plt.ylabel('Marks ')
plt.grid()
plt.show()
#we usually see is data corelate to each other in scatter plot- if proportional or not !
```

```python
roll_no=[1,2,3,4,5,6,7,8]
maths_marks=[60,70,80,76,55,45,34,67]
science_marks=[67,87,56,98,76,89,67,65]
plt.scatter(roll_no,maths_marks,color='red',label='maths')
plt.scatter(roll_no,science_marks,color='blue', label='science')
plt.xlabel('Roll Number of Students')
plt.ylabel('Marks ')
plt.legend()
plt.grid()
plt.show()
```

```python
#Histogram - kinda look like a bar plot - distribution of data but in frequency (tells about range)
```

```python
age=[30,40,50,60,77,80,89,98,76,23,9,10,1]
plt.hist(age,edgecolor='black')
plt.ylabel('Frequency')
plt.xlabel('age')
plt.show()
```

```python
#0 to approx 10-15 its saying there's three people
# from 20 to 30 approx there is 2 people 
#from 40 to 60 there is only 1 and 60 to 70 there only 1 as well
# from 70 to 80 there is about 2
#from 80 to 90 there 1 and 90 to 100 there is 2
#hence define in range which is frequency!
```

```python
#bins - can be change the frequency according to need
age=[30,40,50,60,77,80,89,98,76,23,9,10,1]
plt.hist(age,edgecolor='black',bins=3) # means distribution in 3 
plt.ylabel('Frequency')
plt.xlabel('age')
plt.show()
```

```python
#For Boxplot we use SeaBorn!
import seaborn as sns
#Box Plot --> distribution of data - mostly used to detect outliners 
age=[30,40,50,60,77,80,89,98,76,23,9,10,1]
sns.boxplot(age)
plt.show()
```

```python
#its normal graph we can analyse byself - where 0 to around 20  gives 25%
#20 to approx 50 gives 25% rest and the median is there - which is a mid value
#from median to approx 80 there is 25% percent and end 80 to 100 rest 25%
```

```python
#we can use this data for anomly detection as well
age=[30,40,50,60,77,80,89,98,76,23,9,10,1,300] # we gave the age of 300 which is anomly
sns.boxplot(age)
plt.show() #why plt.show not sns.show 
#Because seaborn has a plt base library - in simple language seaborn is made by plt library so we have to import both
#in case of accessing the seaborn library!
```

```python
#here we can see the data around is normal from 0=100
#the point is anomly here - we can easily analyse it!- so we can remove it accordingly!
```

```python
#Violin plot -> usually if we want to see the distribution of data!
age=[30,40,50,60,77,80,89,98,76,23,9,10,1]
sns.violinplot(age)
plt.show()
```

```python
#from  0 to 25 it getting distributing means more data is there in that range similarly for other things!
```

```python
#plt is a parent library and the sns is child library
#sns is made by plt- we can say it is an advanced version of plt library
#some plots are available in plt some in seaborn 
```

```python
#box and violin are kind of same but in box we use it moreover for outliers(anomaly) - but in box we cant see concentration 
#where as in violen we can see the concentration of data!
age=[30,40,50,60,77,80,89,98,76,23,9,10,1000]
sns.violinplot(age)
plt.show()
```
