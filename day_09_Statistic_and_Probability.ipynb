```python
#Marginal Probability

'''
it is the probability of single event happening without concidering another events
example - you are having a bag full of 8 total balls , 5 balls red - 3 is blue
chance of marginal probabilty of picking a red ball from the total balls

MArginal probability = 5/8
chance of not getting 1-5/8 = 3/8'''
```

```python
#Joint Probability

'''
Probability of two or more event happening in same time

if you roll two dice at a same time , whats prob of getting both of them 3 at same time
first dice --> 1/6 
second dice ---> 1/6

joint probability = 1/6 * 1/6 = 1/36

'''
```

```python
#Conditional Probability

'''
It is the probability of an event happening given that another event is already happens.
# you know , card drawn is face card - porb that the card is king

total possible cards = 52
simply = 4/12 = 1/3
'''
```

```python
#removing outliers 
Q1 = df['Population in thousands (2017)'].quantile(0.25)
Q3 = df['Population in thousands (2017)'].quantile(0.75)
```

```python
IQR = Q3 - Q1
```

```python
LB = Q1-1.5*IQR
UB = Q3+1.5*IQR
```

```python
df = df[(df['Population in thousands (2017)']>LB) & (df['Population in thousands (2017)']<UB)]
```
