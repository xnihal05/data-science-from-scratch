## VIF - Variance Inflation Factor
Checking for multicolinearity - if more than one column(independent) is corelated to each other that affect output

Model gets confused - which is more important
vif>5 high, vif>10 very high

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def vif(df):
    #independent columns
    cols=[]
    for i in df.columns:
        if i!='y':
            cols.append(i)

    data= df[cols]
    print()

#dataframe whic tells the vif of column
    vif_scores=pd.DataFrame()
    vif_scores['Column']=data.columns
    vif_scores['VIF']=0
    print()
#calculate the vif of each column
    v_values=[]
    for i in range(len(cols)):
        v=variance_inflation_factor(data.values,i) #i says please find me vif of first column then 2nd then third
        v_values.append(v)

    vif_scores['VIF']=v_values
    return vif_scores
```

```python
vif(df)
```

```python
def max_vif(df):
    col=df.loc[0]['Column']
    v_max=df.loc[0]['VIF']
    for i in range(len(df)):
        if(df.loc[i]['VIF']>v_max):
            col=df.loc[i]['Column']
            v_max=df.loc[i]['VIF']

    if v_max>5:
        return col
    else:
        return ""
```

```python
def remove_vif(df):
    vif_score= vif(df)
    col=max_vif(vif_score)
    if col=='':
        print('No column with high VIF')
    else:
        print('Removed',col)
        df.drop(col,axis=1,inplace=True)

    print("new VIF Table =",vif(df))
```

## RFE - Recursive Feature Elimination

It's a selection method, used in machine learning, used to choose the most important features for a machine learning

Goal -> Keeps only the feature that really helps the model predict the target, and removee the unimportant features

How does it work?

RFE internally trains and runs the model(Logistic regression) with all the features.

1. It checks and selects the least important feature and removes it.

2. Then it runs the model again ,after removing the least important feature and again finds the least imp feature.

3. Lastly, it gives us a list of all the features and tells us which are the features who are most important.

So we can find the best subset of the data to train our model.

**Why to remove the un-important?**

1) It shifts the focus from quantity to quality.

2) It becomes harder to find patterns when we have too many features.

3) Reduces computational cost, faster training, cheaper deployment.

4) Reduces the chance of **overfitting**

### Confusion Matrix

Tells us how many predictions our model got right or wrong

1. **True Negative** - How many are actual zero and your model also said zero.
2. **False Positive** - How many did model said (1) but actually they were zeros (rediction is positive (1) but actual its zero).
3. **False Negative** - How many did model said (0) but actually they were ones(1)
4. **True Positive** - How many did model said (1) but actually they were ones(1)
