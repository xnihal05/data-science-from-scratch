```python
#a="aabcasaaabcassssbaa" create a dictionary where key is every unique character and value is the number
# of times each unique character is repeting
#eg {a:5,b:7.....}
```

```python
char = 'aabcasaaabcassssbaa'
char_value ={}
for i in char:
    if i not in char_value:
        char_value[i]=1
    else:
        char_value[i]+=1
print(char_value)
```
