```python
#for all the numerical columns show the box plot for outliers

import seaborn as sns
import matplotlib.pyplot as plt

for i in df.columns:
    if (df[i].dtype!= 'object'):
        sns.boxplot(df[i])
        plt.show()
```

```python
#create a function of outlier removal
import numpy as np
def iqr_tech(col):
    Q1 = np.percentile(col,25)
    Q3 = np.percentile(col,75)
    IQR = Q3 - Q1
    LB = Q1 - 1.5*IQR
    UB = Q3 + 1.5*IQR
    return LB,UB
    
```

```python
iqr_tech(df['price'])
```

```python
LB,UB = iqr_tech(df['price'])
```

```python
LB
```

```python
UB
```

```python
df= df[(df['price']>LB) & (df['price']<UB)]
```
