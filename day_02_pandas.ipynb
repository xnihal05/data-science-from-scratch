```python
#if we have null values in dataset we can't go forward with it, it'll give us issues if we want to go in machine learning and all
#Lets deal with null values
# 1. Dropping null values = Removing the null values a
# we cant delete if the dataset is small and precious
# 2. Fill the null values with some values
#Specifically if each and every value is important or dataset is small
```

```python
df_copy = df.copy()
#made copy of dataframe in df_copy
df_copy
```

```python
df_copy.isnull().sum()
```

```python
df_copy.dropna(inplace=True)
```

```python
df_copy.isnull().sum()
#Dropped all the null values
```

```python
#fill the null values
# 1.fill with a value
# 2. fill with mean, median, mode
```

```python
df.fillna(0, inplace=True)
df.isnull().sum()
```

```python
#Filling with mean
df.fillna(df['Bench4Kg'].mean())
#it will fill with all nan value to bench4kg average 
```

```python
#individual mean value filling
df['BodyweightKg']= df['BodyweightKg'].fillna(df['BodyweightKg'].mean())
```
