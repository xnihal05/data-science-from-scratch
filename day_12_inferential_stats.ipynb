```python
#Population standard deviation = 10mg
#population mean value = 500mg
#sample mean = 496mg
#sample size = 50
#is machine working very well or not?


#there is one sample only and it is numerical, std of population is known, mean of both sample
#and population is known- i can run one sample z test 
```

```python
import numpy as np
from scipy import stats
pop_mean=500
pop_std=10
sample_mean=496
sample_size=50
#calculate the z statistic 
z_stat = (sample_mean-pop_mean)/(pop_std/np.sqrt(sample_size)) #formula for z statistics 
#finding p value from z_stat

p_value = stats.norm.sf(abs(z_stat))
print(f'The z stats is = {z_stat:4f}')
print(f'The p value is = {p_value:4f}')

if p_value < 0.01: #states for 95% if 0.05 can be 0.01 for critical values , mean if less than 1% then we'll reject says machine is 
    #wrong - if possibility more than 1% we say there is not enough evidence that machine is wrong
     print('Reject null hypothesis')
else:
    print('failed to reject null')
```
